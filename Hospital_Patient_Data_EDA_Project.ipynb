***Bivariate Analysis (Relationship between 2 variables)***

***1 .Numerical vs Numerical
Example: Age vs Bill Amount
Goal: Check if older patients pay higher bills.***

```python
sns.scatterplot(data=df,x='Age',y='Bill_Amount')
plt.title('Age vs Bill Amount')
plt.show()
print(df['Age'].corr(df['Bill_Amount']))                                                   # correlation
```

***2. Categorical vs Numerical
Example: Department vs Average Bill
Goal: Which department has higher average billing?***

```python
dept_bill = df.groupby('Hospital_Department')['Bill_Amount'].mean()

sns.barplot(x=dept_bill.index, y=dept_bill.values)
plt.title("Department vs Average Bill")
plt.xticks(rotation=45)
plt.show()
```

**3. Categorical vs Categorical
Example: Gender vs Payment Method**

```python
ct = pd.crosstab(df['Gender'], df['Payment_Method'])              #pd.crosstab use for create Frequency table

sns.heatmap(data=ct, annot=True, fmt='d')                         #use annot  for display number on heatmap and fmt use for show number in form of integer
plt.title("Gender vs Payment Method")
plt.xticks(rotation=45)
plt.show()
```

***Convert date column to datetime***

```python
df['Admission_Date']=pd.to_datetime(df['Admission_Date'])
                                                                                     #not show error convert error into NAT(not a time=missing value) 
```

***1. Monthly Admission Trend***

```python
df['Month'] = df['Admission_Date'].dt.month
df.groupby('Month')['Patient_ID'].count().plot(kind='line')                   # when we use pandas plot function then we use kind for decide chart ype
plt.title('Monthly Admission Trend')
plt.show()
```

***2. Weekday Admissions***

```python
df['Weekday'] = df['Admission_Date'].dt.day_name()

df.groupby('Weekday')['Patient_ID'].count().plot(kind='bar')
plt.title('Admissions by Weekday')
plt.xticks(rotation=45)
plt.show()
```

***3. Bill Trend over Time***using line barchat

```python
df.groupby(df['Admission_Date'].dt.to_period('M'))['Bill_Amount'].sum().plot()
plt.title("Monthly Total Billing")
plt.show()
```

***Average stay duration by department***

```python
df['Discharge_Date'] = pd.to_datetime(df['Discharge_Date'], errors='coerce')                    #firstly create stay Durayion Column
df['Stay_Duration'] = (df['Discharge_Date'] - df['Admission_Date']).dt.days
```

```python
dept_stay = df.groupby('Hospital_Department')['Stay_Duration'].mean()                           #Department wise average stay
print(dept_stay)
```

***Correlation of numerial variables***
Check relationship between numbers (Age, Bill Amount, Stay Duration etc)

```python
corr = df[['Age','Bill_Amount','Stay_Duration']].corr()
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

***Insights/findings***

***1) Older patients tend to have slightly higher bills
   2) Cardiology department has highest average stay duration
   3) Admissions peak during winter months
   4) Weekend admissions are lower
   5) Bill amount strongly correlates with stay duration.***
